__*Convert the values of diagnosis into 0 & 1*__

```python
df2=df['diagnosis']
```

```python
df2
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
X=lb.fit(df2)
```

```python
Q=X.fit_transform(df2.values)
```

```python
Q
```

__*Separating the features and target*__

```python
X=df.drop(columns=['diagnosis'])
Y=Q
```

__*Splitting the data into training data & Testing data*__

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

__*Model Training using Logistic Regression*__

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

__*Accuracy Score*__

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
print('Accuracy on score training data = ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print('Accuracy on test data = ', test_data_accuracy)
```
